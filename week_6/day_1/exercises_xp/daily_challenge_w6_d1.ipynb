# Machine Learning – Summary Document

## 1. Introduction to Machine Learning

Machine Learning (ML) is a field of artificial intelligence that allows computers to learn from data and make predictions or decisions without being explicitly programmed.

It is important for data analysts because it helps them:
- Discover patterns in large datasets  
- Make predictions (e.g., sales, risks)  
- Automate decision-making processes  
- Improve business insights  

---

## 2. Applications of Machine Learning

Machine Learning is used in many industries:

- **Finance**: Detecting fraud and predicting loan defaults  
- **Healthcare**: Diagnosing diseases and predicting patient risks  
- **E-commerce**: Recommending products based on user behavior  
- **Transportation**: Optimizing routes and enabling autonomous driving  

---

## 3. Types of Machine Learning

### Supervised Learning
Supervised learning uses labeled data (input + correct output) to train models.

- **Example**: Predicting whether a customer will default on a loan  
- **Type of problem**: Classification or Regression  

---

### Unsupervised Learning
Unsupervised learning works with unlabeled data and finds patterns or groups.

- **Example**: Grouping customers based on purchasing behavior  
- **Type of problem**: Clustering  

---

### Reinforcement Learning
Reinforcement learning is based on learning through interaction with an environment using rewards and penalties.

- **Example**: Training a robot to find the shortest path in a maze  
- **Type of problem**: Decision-making  

---

## 4. Machine Learning Process

### 1. Feature Selection
This step involves choosing the most relevant variables (features) from the dataset that help the model make accurate predictions.

Example: Selecting income, credit score, and debt ratio for loan prediction.

---

### 2. Model Selection
In this step, we choose the appropriate algorithm based on the problem.

Examples:
- Logistic Regression for classification  
- K-Means for clustering  
- Q-Learning for reinforcement learning  

---

### 3. Model Evaluation
This step measures how well the model performs on unseen data.

Common metrics:
- Accuracy, Precision, Recall, F1-score (for classification)  
- Silhouette Score (for clustering)  
- Cumulative reward (for reinforcement learning)  

---

## Machine Learning Workflow

Data Collection → Data Cleaning & Preprocessing → Feature Selection →  
Model Selection → Model Training → Model Evaluation →  
Model Optimization (Tuning) → Final Model / Deployment
